### Adding all months data in one file and convert it into one csv file

```python
files = [file for file in os.listdir('./Sales_Data')]

all_months_data = pd.DataFrame()

for file in files:
    df = pd.read_csv("./Sales_Data/"+file)
    all_months_data = pd.concat([all_months_data, df])
    
all_months_data.to_csv("all_data.csv", index=False)
```

```python
df = pd.read_csv('all_data.csv')
df.shape
```

### Removing all nan values from the dataset

```python
nan = df[df.isna().any(axis=1)]
nan

df = df.dropna(how='all')
df
```

```python
df
```

### Finding Or and Delete it

```python
df = df[df['Order Date'].str[0:2] != 'Or']
```

### Convert column to the correct type

```python
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'])  #make int
df['Price Each'] = pd.to_numeric(df['Price Each'])  #make float
df.head()
```

### Adding Month Column at the end

```python
df['Month'] = df['Order Date'].str[0:2]
df
```

```python
df['Month'] = df['Month'].astype('int32')
```

```python
df
```

### Add a Sales Column 

```python
df['Sales'] = df['Quantity Ordered'] * df['Price Each']
df
```

### Add City column 

```python
def get_city(address):
    return address.split(',')[1]


df['City'] = df['Purchase Address'].apply(lambda x : get_city(x))
df
```

### Add state with city column

```python
def get_city(address):
    return address.split(',')[1]

def get_State(address):
    return address.split(',')[2].split(' ')[1]

df['City'] = df['Purchase Address'].apply(lambda x : f"{get_city(x)} ({get_State(x)})")
df
```

```python
# What are the top 5 most common products ordered together? (Hint: Look for duplicate 'Order ID' with different 'Product' values)

from collections import Counter
import itertools

# Group the data by 'Order ID' and aggregate the products ordered in each order
grouped_orders = df.groupby('Order ID')['Product'].apply(list)

# Create a list to store combinations of products
product_combinations = []

# Iterate through each order and find combinations of products
for order_products in grouped_orders:
    product_combinations.extend(list(itertools.combinations(order_products, 2)))

# Count the occurrences of each product combination
combination_counts = Counter(product_combinations)

# Find the top 5 most common product combinations
top_combinations = combination_counts.most_common(5)

# Print the top 5 most common product combinations
for i, combination in enumerate(top_combinations, 1):
    products, count = combination
    print(f"{i}. Products: {products}, Count: {count}")
```

```python
# graph representation of above cell

# Extract the product combinations and their counts
top_combinations_products = [combination[0] for combination in top_combinations]
top_combinations_counts = [combination[1] for combination in top_combinations]

# Plot the bar chart
plt.figure(figsize=(10, 6))
plt.bar(range(len(top_combinations_products)), top_combinations_counts, color='skyblue')
plt.xlabel('Product Combinations')
plt.ylabel('Number of Orders')
plt.title('Top 5 Most Common Products Ordered Together')
plt.xticks(range(len(top_combinations_products)), [', '.join(products) for products in top_combinations_products], rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# What is the trend of sales over the months?

# Group the data by 'Month' and calculate total sales revenue for each month
monthly_sales = df.groupby('Month')['Sales'].sum()

# Plot the line chart
plt.figure(figsize=(10, 6))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o', color='green', linestyle='-')
plt.title('Trend of Sales Over the Months')
plt.xlabel('Month')
plt.ylabel('Total Sales Revenue ($)')
plt.xticks(monthly_sales.index)
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Machine Learning model to detect the outliers

```python
# Are there any outliers in the data? 

from sklearn.ensemble import IsolationForest

# Fit Isolation Forest model
clf = IsolationForest(contamination=0.05)  # Adjust contamination parameter as needed
clf.fit(df[['Sales']])

# Predict outliers
outliers = clf.predict(df[['Sales']])

# Add outlier predictions to the DataFrame
df['Outlier'] = outliers

# Visualize outliers
plt.figure(figsize=(10, 6))
plt.scatter(df.index, df['Sales'], c=df['Outlier'], cmap='viridis')
plt.title('Outliers Detected in Sales Data')
plt.xlabel('Index')
plt.ylabel('Sales')
plt.colorbar(label='Outlier Prediction')
plt.tight_layout()
plt.show()
```
